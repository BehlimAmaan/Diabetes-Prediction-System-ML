```python
df = pd.read_csv("D:\Projects\Ml project\Fraud Prediction\data\Raw\diabetes_prediction_dataset.csv")
df.head()
```

# Handling Outliers

```python
Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['bmi'] < lower) | (df['bmi'] > upper)]
print(outliers)
```

```python
df['bmi'] = df['bmi'].clip(upper, lower)
```

```python
Q1 = df['blood_glucose_level'].quantile(0.25)
Q3 = df['blood_glucose_level'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['blood_glucose_level'] < lower) | (df['blood_glucose_level'] > upper)]
print(outliers)
```

```python
df['blood_glucose_level'] = df['blood_glucose_level'].clip(upper, lower)
```

# Encoding Catogorical Features 

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown = 'ignore')
encoded = encoder.fit_transform(df[['gender','smoking_history']]).toarray()

encoded_df = pd.DataFrame(
    encoded, columns=encoder.get_feature_names_out(['gender', 'smoking_history'])
)

df_encoded = pd.concat([df.drop(['gender', 'smoking_history'], axis=1), encoded_df], axis=1)
```
